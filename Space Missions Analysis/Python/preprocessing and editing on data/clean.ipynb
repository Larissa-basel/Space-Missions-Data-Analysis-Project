```python
import pycountry
df=pd.read_csv("Data/agencies.csv")
def country_to_alpha3(name):
    try:
        return pycountry.countries.lookup(name).alpha_3
    except:
        return None

df["Country"] = df["Country"].apply(country_to_alpha3)
df = df.dropna(subset=["Country"])
```

```python
df.to_csv("../../Data/agencies.csv", index=False) 
```

```python
import pandas as pd
import numpy as np

# Load your dataframe
df = pd.read_csv("Data/spacecraft.csv")

K = 504                
N = len(df)            
base = N // K          
rem  = N % K          

counts = np.full(K, base, dtype=int)


if rem > 0:
    idxs = np.linspace(0, K - 1, rem, endpoint=False).round().astype(int)
    counts[idxs] += 1

ids = []
for i, c in enumerate(counts, start=1):
    ids.extend([i] * c)
ids = ids[:N]

df["Mission_id"] = ids

print(f"Total rows: {N}, unique IDs: {df['Mission_id'].nunique()}, remainder distributed: {rem}")
df.head(), df.tail()
```

```python
df.to_csv("Data/spacecraft.csv", index=False)
```

```python
import pandas as pd


df = pd.read_csv("Data/mission_crew.csv")

num_missions = 504
num_rows = len(df)


mission_ids = []
mission_id = 1

for i in range(num_missions):
    if i % (num_missions // 294) == 0 and i < 294:  
        mission_ids.extend([mission_id] * 4)
    else:
        mission_ids.extend([mission_id] * 3)
    mission_id += 1


mission_ids = mission_ids[:num_rows]


df["Mission_id"] = mission_ids

df.to_csv("Data/mission_crew.csv", index=False)

```
