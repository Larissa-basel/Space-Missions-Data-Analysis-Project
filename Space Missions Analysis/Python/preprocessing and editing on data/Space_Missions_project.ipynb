```python
import pandas as pd
import random
from faker import Faker
```

```python
fake = Faker()
# Parameters
num_agencies = 5
num_missions = 60
num_spacecraft = 80
num_astronauts = 50
num_mission_crew = 80
```

```python
# Agencies
agency_names = ["NASA", "ESA", "SpaceX", "Roscosmos", "ISRO"]
countries = ["USA", "Europe", "USA", "Russia", "India"]
founded_years = [1958, 1975, 2002, 1992, 1969]
budgets = [22.6, 6.7, 3.2, 4.1, 1.9]

agencies = pd.DataFrame({
    "agency_id": range(1, num_agencies + 1),
    "agency_name": agency_names,
    "country": countries,
    "founded_year": founded_years,
    "budget_billion_usd": budgets
})
```

```python
# Missions
destinations = ["Moon", "Mars", "ISS", "Jupiter", "Saturn", "Venus","Low Earth Orbit"]
mission_types = ["Manned", "Unmanned", "Satellite Deployment"]
statuses = ["Completed", "Failed"]

missions = []
for i in range(1, num_missions + 1):
    agency_id = random.randint(1, num_agencies)
    missions.append({
        "mission_id": i,
        "mission_name": f"Mission {i}",
        "agency_id": agency_id,
        "launch_date": fake.date_between(start_date="-30y", end_date="-2y"),
        "destination": random.choice(destinations),
        "mission_type": random.choice(mission_types),
        "status": random.choice(statuses),
        "duration_days": random.randint(20, 400)
    })

missions_df = pd.DataFrame(missions)
```

```python
#  Spacecraft
types = ["Rocket", "Shuttle", "Rover"]
manufacturers = ["Boeing", "Lockheed Martin", "SpaceX", "Blue Origin", "Roscosmos","Airbus", "ISRO"]

spacecraft = []
for i in range(1, num_spacecraft + 1):
    mission_id = random.randint(1, num_missions)
    spacecraft.append({
        "spacecraft_id": i,
        "name": f"Spacecraft {i}",
        "type": random.choice(types),
        "cost_million_usd": round(random.uniform(50, 500), 2),
        "manufacturer": random.choice(manufacturers),
        "first_launch_year": random.randint(1960, 2022),
        "mission_id": mission_id
    })

spacecraft_df = pd.DataFrame(spacecraft)
```

```python
# Astronauts
astronauts = []
for i in range(1, num_astronauts + 1):
    astronauts.append({
        "astronaut_id": i,
        "name": fake.name(),
        "nationality": fake.country(),
        "gender": random.choice(["Male", "Female"]),
        "birth_year": random.randint(1940, 2000),
        "missions_count": random.randint(1, 5)
    })

astronauts_df = pd.DataFrame(astronauts)
```

```python
# Mission_Crew
mission_crew = []
for i in range(1, num_mission_crew + 1):
    mission_id = random.randint(1, num_missions)
    astronaut_id = random.randint(1, num_astronauts)
    mission_crew.append({
        "mission_id": mission_id,
        "astronaut_id": astronaut_id,
        "role": random.choice(["Commander", "Pilot", "Engineer", "Scientist"])
    })

mission_crew_df = pd.DataFrame(mission_crew)
```
